```python
# Drop unnecessary columns
news = news.drop(columns=['relevancyScore'])
```

```python
# Convert articlePublishedDate to datetime objects
news['articlePublishedDate'] = pd.to_datetime(news['articlePublishedDate'])

# Extract just the date part
news['date'] = news['articlePublishedDate'].dt.date

# Define the market close time
market_close_time = time(16, 0, 0)

# Find articles published closest to the market close time
news['time_diff'] = news['articlePublishedDate'].apply(lambda x: abs((x - x.replace(hour=market_close_time.hour, minute=market_close_time.minute, second=market_close_time.second)).total_seconds()))
news = news.sort_values(['date', 'time_diff']).groupby('date', as_index=False).first()

# Drop the 'time_diff' and 'date' columns as they are no longer needed
news = news.drop(columns=['time_diff', 'articlePublishedDate'])
```

```python
import requests
from datetime import datetime, timedelta
import time
from tqdm import tqdm

def paginate_news_requests(api_key, ticker, start_time, end_time, limit=200):
    base_url = "https://www.alphavantage.co/query"
    all_articles = []
    time_from = start_time
    time_to = end_time
    requests_count = 0
    max_daily_requests = 500
    delay_between_requests = 12  # 5 requests per minute

    with tqdm(total=max_daily_requests) as pbar:
        while time_from < time_to and requests_count < max_daily_requests:
            response = requests.get(base_url, params={
                "function": "NEWS_SENTIMENT",
                "tickers": ticker,
                "time_from": time_from.strftime("%Y%m%dT%H%M"),
                "time_to": time_to.strftime("%Y%m%dT%H%M"),
                "limit": limit,
                "apikey": api_key
            })

            requests_count += 1
            pbar.update(1)  # Update the progress bar
            data = response.json()
            articles = data.get('feed', [])

            if not articles:
                break

            all_articles.extend(articles)
            last_article_time = datetime.strptime(articles[-1]['time_published'], "%Y%m%dT%H%M%S")
            time_to = last_article_time - timedelta(seconds=1)

            if time_from < time_to and requests_count < max_daily_requests:
                time.sleep(delay_between_requests)

    return all_articles

# Usage example
api_key = "03Y7A0BO5DBI8C5L"
ticker = "AAPL"
start_time = datetime(2022, 4, 1, 0,0)
end_time = datetime(2022, 4, 23, 23, 59)

articles = paginate_news_requests(api_key, ticker, start_time, end_time)
```

```python
def save_to_dataframe(articles):
    data = []

    for article in articles:
        data.append({
            "articleHeadline": article["title"],
            "articleURL": article["url"],
            "time_published": article["time_published"],
            "summary": article["summary"],
            'overall_sentiment_score': article['overall_sentiment_score'],
            'overall_sentiment_label': article['overall_sentiment_label'],
        })

    df = pd.DataFrame(data)
    return df

# Usage example
df = save_to_dataframe(articles)
```

```python
# Convert articlePublishedDate to datetime objects
df['time_published'] = pd.to_datetime(df['time_published'])

# Extract just the date part
df['date'] = df['time_published'].dt.date

# Define the market close time
market_close_time = time(16, 0, 0)

# Find articles published closest to the market close time
df['time_diff'] = df['time_published'].apply(lambda x: abs((x - x.replace(hour=market_close_time.hour, minute=market_close_time.minute, second=market_close_time.second)).total_seconds()))
df = df.sort_values(['date', 'time_diff']).groupby('date', as_index=False).first()

# Drop the 'time_diff' and 'date' columns as they are no longer needed
df = df.drop(columns=['time_diff', 'time_published'])
```
